```python
import pandas as pd
import random

# Load the CSV file with a semicolon delimiter
file_path = "/content/bank-full.csv"
df = pd.read_csv(file_path, delimiter=';')

# List of random US states (you can modify this list as needed)
states = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado',
          'Connecticut', 'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho',
          'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana',
          'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota',
          'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire',
          'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota',
          'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina',
          'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia',
          'Washington', 'West Virginia', 'Wisconsin', 'Wyoming']

# Add a 'State' column with random states to the DataFrame
df['State'] = [random.choice(states) for _ in range(len(df))]
```

```python
df.rename(columns = {
                      'housing':'housing_loan',
                      'loan':'personal_loan',
                      'duration':'duration(s)',
                      'campaign':'curr_camp_contacts',
                      'previous':'prev_camp_contacts',
                      'poutcome':'prev_camp_outcome',
                      'y':'subscribed'
                      },inplace=True)
```

```python
# Dictionary to map month names to numeric values
month_mapping = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}

# Replace the month names with numeric values
df['month'] = df['month'].replace(month_mapping)

# Combining the day and month column to get the desired date stamp for further analysis
df['date_of_last_contact'] = df.apply(lambda row: f"{row['day']}/{row['month']}/2023", axis=1,inplace=True)
df.drop(['day', 'month'], axis=1, inplace=True)
```
